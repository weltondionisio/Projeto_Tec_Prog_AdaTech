```python
cadastrados = [['Naruto', '12345678', 'Vila da Folha', 1, True], ['Gaara', '87654321', 'Vila da Areia', 2, True],
         ['Luffy', '30567892', 'Vila Foosha', 3, True], ['Itadori', '09876543', 'Sendai', 4, True],
         ['Goku', '12139785', 'Capital do Oeste', 5, True]]

identificador = 5
comando = 1

while comando == 1 or comando <=6:
  comando = int(input("Bem-vindo ao nosso sistema! O que deseja? \n1 - Inserir usuário \n2 - Excluir usuário\n3 - Atualizar usuário \n4 - Informações de um usuário \n5 - Informações de todos os usuários \n6 - Sair\n"))
  if comando == 1:
    nome = input("Insira o nome do usuário: ")
    telefone = input('Insira o telefone do usuário: ')
    endereco = input('Insira o endereço do usuário: ')
    identificador += 1
    cadastrados.append([nome, telefone, endereco, identificador, True])
    print("\nO novo usuário foi cadastrado \n")
      
  elif comando == 2:
    delimitador = 1
    while delimitador == 1:
      apagar = int(input("Insira o id do usuário: "))
      for usuarios in cadastrados:  
        if usuarios[3] != apagar:
          delimitador = 1
        else:
          usuarios[4] = False
      for usuarios in cadastrados:
        if usuarios[3] == apagar:
          delimitador = 0
          print('O usuário foi excluído.')
      if delimitador == 1:
        print('Usuário não encontrado!')
  
  elif comando == 3:
    opcao = 0
    delimitador_b = 1
    while delimitador_b == 1:
      atualizar = int(input("Insira  id do usuário: "))
      for usuarios in cadastrados:  
        if usuarios[3] != atualizar:
          delimitador_b = 1
        else:
          delimitador_b = 0
          break
      if delimitador_b == 1:
        print("Usuário não encontrado!")    
      while delimitador_b == 0:
        opcao = int(input('''Qual informação deseja alterar?
                1 - Nome
                2 - Telefone
                3 - Endereço'''))
        if opcao == 1:
          novo_nome = input('Insira o nome: ')
          for usuarios in cadastrados:
            usuarios[0] = novo_nome
            break
        elif opcao == 2:
          novo_telefone = input('Insira o telefone: ')
          usuarios[1] = novo_telefone
          break
        elif opcao == 3:
          novo_endereco = input('Insira o endereço: ')
          usuarios[2] = novo_endereco
          break
        else:
          continue
    continue 
        
  elif comando == 4:
    delimitador_c = 1
    while delimitador_c == 1:
      atualizar = int(input("4 \nInsira o ID do usuário:"))
      for usuarios in cadastrados:  
        if usuarios[3] != atualizar:
          delimitador_c = 1
        else:
          delimitador_c = 0
          break
      if delimitador_c == 1:
        print("Usuário não encontrado!")
    while delimitador_c == 0:
      for usuarios in cadastrados:
        if usuarios[3] == atualizar:
          print(f'Nome: {usuarios[0]} \nTelefone: {usuarios[1]} \nEndereço: {usuarios[2]}\n')
          break
      break
    continue

  elif comando == 5:
    nome = []
    telefone = []
    endereco = []
    id_usuario = []
    id_ativo = []
    for usuarios in cadastrados:
        nome.append(usuarios[0])
        telefone.append(usuarios[1])
        endereco.append(usuarios[2])
        id_usuario.append(usuarios[3])
        id_ativo.append(usuarios[4])
    for i in range(len(id_usuario)):
      print(f'ID: {id_usuario[i]} \nNome: {nome[i]} \nTelefone: {telefone[i]} \nEndereço: {endereco[i]} \nRegistro ativo: {id_ativo[i]}\n')
  
  elif comando == 6:
    print('O programa foi encerrado.')
    break
  else:
    print('Selecione uma opção válida.')
    comando = 1
```
